```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
poblacion = pd.read_csv('./DataSets/PoblacionEspaña.csv', sep=';', low_memory=False, decimal=",")
poblacion
```

```python
#Separamos el año
poblacion['Anyo'] = poblacion['Periodo'].str.replace('1 de enero de ','').str.replace('1 de julio de ','')
#Separamos el mas y adecuamos a formato
poblacion['Mes'] = poblacion['Periodo'].str.replace('1 de enero de ','M01').str.replace('1 de julio de ','M07').apply(lambda x: x[:-4])
#Unificamos
poblacion['Periodo'] = poblacion['Anyo']+poblacion['Mes']
#Eliminamos columnas temporales
poblacion = poblacion.drop(['Mes'], axis=1)
poblacion = poblacion.drop(['Anyo'], axis=1)
poblacion
```

```python
#Cambiamos nombre
poblacion["TotalPoblación"] = poblacion["Total"]
poblacion = poblacion.drop(['Total'], axis=1)

poblacion['TotalPoblación'] = poblacion['TotalPoblación'].str.replace('.','')

poblacion['TotalPoblación'] = poblacion['TotalPoblación'].astype('float')

poblacion
```

```python
#Sacamos solamente ambos sexos y eliminamos la columna
poblacion = poblacion.loc[poblacion['Sexo']=='Ambos sexos']
poblacion = poblacion.drop(['Sexo'], axis=1)
poblacion
```

```python
#Sacamos solamente los de todas las edades
poblacion = poblacion.loc[poblacion['Edad']=='Total']
poblacion = poblacion.drop(['Edad'], axis=1)
poblacion
```

```python
print(poblacion.dtypes)
poblacion
```

```python
CompraventaViviendas = pd.read_csv('./DataSets/CompraventaViviendasEspaña.csv', sep=';', low_memory=False, decimal="," )
CompraventaViviendas
```

```python
#sacamos solo el total de viviendas compradas/vendidas , no nos interesan distinguir entre vivienda nueva, usada, libre o protegida 
CompraventaViviendas = CompraventaViviendas.loc[CompraventaViviendas['Régimen y estado']=='Viviendas: Total']
CompraventaViviendas = CompraventaViviendas.drop(['Régimen y estado'], axis=1)
CompraventaViviendas
```

```python
CompraventaViviendas['Comunidades y Ciudades Autónomas'] = CompraventaViviendas['Comunidades y Ciudades Autónomas'].fillna('Total Nacional')
CompraventaViviendas['Provincias'] = CompraventaViviendas['Provincias'].fillna('Total Nacional')
CompraventaViviendas
```

```python
CompraventaViviendas["NumCompraVentas"] = CompraventaViviendas["Total"]
CompraventaViviendas = CompraventaViviendas.drop(['Total'], axis=1)

CompraventaViviendas['NumCompraVentas'] = CompraventaViviendas['NumCompraVentas'].str.replace('.','')
CompraventaViviendas['NumCompraVentas'] = CompraventaViviendas['NumCompraVentas'].astype('float')
CompraventaViviendas
print(CompraventaViviendas.dtypes)
```

```python
CompraventaViviendas = CompraventaViviendas.drop(['Total Nacional'], axis=1)

#Reseteamos el index
CompraventaViviendas.reset_index(drop = True , inplace = True )
CompraventaViviendas
```

```python
poblacion
poblacionTotal = poblacion
poblacionProvincias = poblacion
```

```python
#Saco solo los totales
poblacionTotal = poblacionTotal.loc[poblacionTotal['Provincias']=='Total Nacional']
poblacionTotal['Comunidades y Ciudades Autónomas']=poblacionTotal['Provincias']
poblacionTotal= poblacionTotal.drop('Provincias', axis=1)
poblacionTotal
```

```python
#Saco solo las provincias
poblacionProvincias = poblacionProvincias.loc[poblacionProvincias['Provincias']!='Total Nacional']
poblacionProvincias
```

```python
CompraventaViviendas
```

```python
#Hacemos merge por TOTALES
poblacionYcompraventaTOTAL = pd.merge(CompraventaViviendas,poblacionTotal, how = "inner")

print(poblacionYcompraventaTOTAL.dtypes)
poblacionYcompraventaTOTAL
```

```python
#Hacemos merge por PROVINCIAS
poblacionYcompraventaPROVINCIAS = pd.merge(CompraventaViviendas,poblacionProvincias, how = "inner" )
print(poblacionYcompraventaPROVINCIAS.dtypes)
poblacionYcompraventaPROVINCIAS
```

```python
poblacionYcompraventa = pd.concat([poblacionYcompraventaTOTAL, poblacionYcompraventaPROVINCIAS], axis=0)
poblacionYcompraventa
```

```python
import pandas as pd
import numpy as np
viviendas = pd.read_csv('./DataSets/PrecioViviendasEspaña.csv', sep=';', low_memory=False, decimal=",")
viviendas
```

```python
#sacamos solo el dato de indice, no nos interesan las variaciones 
viviendas = viviendas.loc[viviendas['Índices y tasas']=='Índice']
viviendas = viviendas.drop(['Índices y tasas'], axis=1)

#sacamos solo el dato general, no nos interesan distinguir entre vivienda nueva o de segunda mano 
viviendas = viviendas.loc[viviendas['General, vivienda nueva y de segunda mano']=='General']
viviendas = viviendas.drop(['General, vivienda nueva y de segunda mano'], axis=1)
viviendas = viviendas.drop(['Total Nacional'], axis=1)

viviendas
```

```python
#Eliminamos la columa Total nacional y las filas con algun valor nulo (LAS PROVINCIAS)
viviendas['Comunidades y Ciudades Autónomas'] = viviendas['Comunidades y Ciudades Autónomas'].fillna('Total Nacional')

viviendas
```

```python
viviendas["PrecioVivienda"] = viviendas["Total"]
viviendas = viviendas.drop(['Total'], axis=1)

#Cambio formato periodo para adapatar al resto dataSets
viviendas['Periodo'] = viviendas['Periodo'].str.replace('T1','M01')
viviendas['Periodo'] = viviendas['Periodo'].str.replace('T2','M04')
viviendas['Periodo'] = viviendas['Periodo'].str.replace('T3','M07')
viviendas['Periodo'] = viviendas['Periodo'].str.replace('T4','M10')
viviendas
```

```python
viviendas['PrecioVivienda'] = viviendas['PrecioVivienda']*1000
print(viviendas.dtypes)

#Reseteamos el indice
viviendas.reset_index(drop = True , inplace = True )

viviendas
```
